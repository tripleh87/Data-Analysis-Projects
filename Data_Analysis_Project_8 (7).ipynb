**4.2 Gross sales vs based on specific manufacturer, and year**

**4.2.1 Plotly**

**4.2.2 Matplotlib**

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming your dataframe is named 'dataframe'
# Extract necessary columns
gross_sales = dataframe["Gross Sales"]
discounts = dataframe["Discounts"]
manufacturer = dataframe["Product"]
year = dataframe["Year"]

# Create a scatter plot
fig, ax = plt.subplots(figsize=(10, 6))  # Adjust figure size as needed

# Define a list of manufacturers to highlight
manufacturers_to_highlight = ["Manufacturer1", "Manufacturer2"]  # Replace with your desired manufacturers

# Plot data for highlighted manufacturers with distinct colors and labels
for manufacturer_name in manufacturers_to_highlight:
    manufacturer_data = dataframe[dataframe["Product"] == manufacturer_name]
    ax.scatter(
        manufacturer_data["Gross Sales"],
        manufacturer_data["Discounts"],
        label=manufacturer_name,
        alpha=0.7,
    )

other_data = dataframe[~dataframe["Product"].isin(manufacturers_to_highlight)]
ax.scatter(
    other_data["Gross Sales"], other_data["Discounts"], color="gray", label="Other", alpha=0.5
)

# Set plot title and labels
ax.set_title("Gross Sales vs Discounts by Manufacturer and Year")
ax.set_xlabel("Gross Sales (in millions)")
ax.set_ylabel("Discounts")

# Add a colorbar to represent the year
scatter = ax.scatter(
    gross_sales, discounts, c=year, cmap="viridis", alpha=0.7  # Choose a suitable colormap
)
colorbar = fig.colorbar(scatter)
colorbar.set_label("Year")

# Add a legend
ax.legend()
plt.show()
```
