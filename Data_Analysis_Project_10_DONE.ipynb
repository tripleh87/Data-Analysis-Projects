## **2. Detailed data analysis of self chosen columns.**

**Let us now have all the columns and think what data we can compare.**

```python
columns = dataframe.columns
print(columns)
```

**2.1 We are going to compare the following data:  Operational schools (Operational_schools) by state (Statename)**

**Matplotlib histogram.**

```python
state_school_counts = dataframe.groupby('STATENAME')['OPERATIONAL_SCHOOLS'].sum()

# Create the histogram
plt.figure(figsize=(12, 6))  # Adjust figure size as needed
plt.bar(state_school_counts.index, state_school_counts.values)
plt.xlabel('State')
plt.ylabel('Number of Operational Schools')
plt.title('Distribution of Operational Schools by State')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

**Plotly histogram**

```python
px.histogram(dataframe, x='STATENAME', y='OPERATIONAL_SCHOOLS', title='Distribution of Operational Schools by State')
```

**2.2 Lets see how LEAID and FIPST are correlating between each other. For this we are going to use a scatter plot with Matplotlib and Plotly.**

**Matplotlib scatter plot.**

```python
leaid = dataframe['LEAID']
fipst = dataframe['FIPST']

plt.scatter(leaid, fipst, color = "red")
plt.xlabel('LEAID (in millions)')
plt.ylabel('FIPST')
plt.title('Relationship between LEAID and FIPST')
plt.show()
```

**Plotly scatter plot.**

```python
px.scatter(dataframe, x='LEAID', y='FIPST', title='Relationship between LEAID and FIPST')
```

**Lets have a reminder of the columns we have, to see what data we can use in order to plot it using a pie chart, both with matplotlib and plotly.**

```python
columns = dataframe.columns
print(columns)
```

**2.3 Let us now see a pie chart using both matplotlib and plotly for state (STATENAME) and operational schools (OPERATIONAL_SCHOOLS)**

**Matplotlib pie chart**

```python
state = dataframe['STATENAME']
operational_schools = dataframe['OPERATIONAL_SCHOOLS']

# Group data by state and sum operational schools
state_school_counts = dataframe.groupby('STATENAME')['OPERATIONAL_SCHOOLS'].sum()

# Create the pie chart
plt.figure(figsize=(10, 10))  # Adjust figure size as needed
plt.pie(state_school_counts, labels=state_school_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Operational Schools by State')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

**Plotly pie chart**

```python
fig = px.pie(dataframe, values='OPERATIONAL_SCHOOLS', names='STATENAME', title='Distribution of Operational Schools by State')
fig.update_layout(
    width=1000,
    height=1000)
fig.show()
```

**2.4 For next analysis we will use: Operational schools (OPERATIONAL_SCHOOLS) and Lzip4 (LZIP4)**

**Matplotlib histogram**

```python
operational_schools = dataframe['OPERATIONAL_SCHOOLS']
lzip4 = dataframe['LZIP4']

plt.hist(operational_schools, bins=20, alpha=0.5, label='Operational Schools')  # Adjust bins as needed
plt.hist(lzip4, bins=20, alpha=0.5, label='LZIP4')  # Adjust bins as needed
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Distribution of Operational Schools and LZIP4')
plt.legend(loc='upper right')
plt.show()
```

**Plotly histogram**

```python
px.histogram(dataframe, x='OPERATIONAL_SCHOOLS', y='LZIP4', title='Relationship between OPERATIONAL_SCHOOLS and LZIP4')
```
