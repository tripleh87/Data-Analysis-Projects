Now let's see how looks the distribution of values between Milk and Sugar for Muffins

```python
muffin_milk = recipies[recipies['Type'] == 'Muffin']['Milk']
muffin_sugar = recipies[recipies['Type'] == 'Muffin']['Sugar']
cupcake_milk = recipies[recipies['Type'] == 'Cupcake']['Milk']
cupcake_sugar = recipies[recipies['Type'] == 'Cupcake']['Sugar']
plt.scatter(muffin_milk, muffin_sugar, color = "blue", label = "Muffin values")
plt.scatter(cupcake_milk,cupcake_sugar, color = "red", label = "Cupcake values")
plt.title("Distribution of milk and sugar in muffins and cupcakes")
plt.xlabel("How much milk")
plt.ylabel("How much sugar")
plt.legend(loc = "upper left")
```

Now lets see how are the values for egg and butter in both muffins and cupcakes

```python
muffin_egg = recipies[recipies['Type'] == 'Muffin']['Egg']
muffin_butter = recipies[recipies['Type'] == 'Muffin']['Butter']
cupcake_egg = recipies[recipies['Type'] == 'Cupcake']['Egg']
cupcake_butter = recipies[recipies['Type'] == 'Cupcake']['Butter']

plt.scatter(muffin_egg, muffin_butter, color = "green", label = "Muffin butter and egg")
plt.scatter(cupcake_egg, cupcake_butter, color = "orange", label = "Cupcake butter and egg")
plt.title("Distribution of egg and butter for muffins and cupcakes")
plt.xlabel("Egg values")
plt.ylabel("Butter values")
plt.legend(loc = 'upper left')
```
